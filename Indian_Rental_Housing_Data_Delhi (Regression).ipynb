```python
# Let us assign X and y variables
X = df[['house_type', 'house_size', 'location', 'city', 'numBathrooms', 'SecurityDeposit', 'Status']]
y = df['price']
```

```python
# Splitting the data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                   test_size = 0.2,
                                                   random_state = 1)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train.dtypes
```

```python
# Splitting numerical and categorical columns seperately
X_train_num = X_train.select_dtypes(include = ['int32', 'float64'])
X_test_num = X_test.select_dtypes(include = ['int32', 'float64'])
print(X_train_num.shape)
print(X_test_num.shape)
```

```python
X_train_cat = X_train.select_dtypes(include = ['object'])
X_test_cat = X_test.select_dtypes(include = ['object'])
print(X_train_cat.shape)
print(X_test_cat.shape)
```

```python
# Scaling the numerical data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_num_trans = sc.fit_transform(X_train_num)
X_test_num_trans = sc.transform(X_test_num)
print(X_train_num_trans.shape)
print(X_test_num_trans.shape)
```

```python
# Scaling the categorical data
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
X_train_cat_trans = oe.fit_transform(X_train_cat)
X_test_cat_trans = oe.transform(X_test_cat)
print(X_train_cat_trans.shape)
print(X_test_cat_trans.shape)
```

```python
X_train_transformed = np.concatenate((X_train_num_trans, X_train_cat_trans), axis=1)
X_test_transformed = np.concatenate((X_test_num_trans, X_test_cat_trans), axis=1)
print(X_train_transformed.shape)
print(X_test_transformed.shape)
```

```python
X_train_transformed = pd.DataFrame(X_train_transformed)
X_test_transformed = pd.DataFrame(X_test_transformed )
print(X_train_transformed.shape)
print(X_test_transformed.shape)
```

```python
%%time
# Building the model
# KNeighbors
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn.fit(X_train_transformed, y_train)
```

```python
%%time  
# Let us apply test data on model
y_pred = knn.predict(X_test_transformed)
print(len(y_pred))
print(len(y_test))
```

```python
training_time = []
prediction_time = []
mse = []
rmse = []
mae = []
r2= []
```

```python
# Evaluating model
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
# Evaluating model
from sklearn.metrics import root_mean_squared_error
root_mean_squared_error(y_test, y_pred)
```

```python
# Evaluating model
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
```

```python
# Evaluating model
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
